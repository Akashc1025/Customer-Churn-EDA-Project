```python
ax=sns.countplot(x=df["SeniorCitizen"],data=df)
ax.bar_label(ax.containers[0])
plt.title("count of customer by churn")
```

```python




# Step 1: Group by SeniorCitizen (Yes/No) and Churn
counts = df.groupby(["SeniorCitizen", "Churn"]).size().reset_index(name="count")

# Step 2: Convert to percentage
counts["percentage"] = counts.groupby("SeniorCitizen")["count"].transform(lambda x: x / x.sum() * 100)

# Step 3: Pivot for stacked bar
pivot = counts.pivot(index="SeniorCitizen", columns="Churn", values="percentage").fillna(0)

# Step 4: Plot stacked bar with Yes/No on X-axis
ax = pivot.plot(kind="bar", stacked=True, figsize=(6,5), color=['#1f77b4','#ff7f0e'])

# Step 5: Add percentage labels
for container in ax.containers:
    ax.bar_label(container, fmt="%.1f%%", label_type="center")

plt.title("Senior Citizen vs Churn (Percentage Stacked)")
plt.ylabel("Percentage")
plt.xlabel("Senior Citizen")
plt.legend(title="Churn")
plt.show()

```

```python
ax=sns.countplot(x=df["Contract"],data=df)
ax.bar_label(ax.containers[0])
plt.title("count of customer by Contract")
```

```python
counts = df.groupby(["Contract", "Churn"]).size().reset_index(name="count")

# Step 2: Convert to percentage
counts["percentage"] = counts.groupby("Contract")["count"].transform(lambda x: x / x.sum() * 100)

# Step 3: Pivot for stacked bar
pivot = counts.pivot(index="Contract", columns="Churn", values="percentage").fillna(0)

# Step 4: Plot stacked bar
ax = pivot.plot(kind="bar", stacked=True, figsize=(7,5), color=['#1f77b4','#ff7f0e'])

# Step 5: Add percentage labels inside bars
for container in ax.containers:
    ax.bar_label(container, fmt="%.1f%%", label_type="center")

plt.title("Churn Distribution by Contract (Percentage Stacked)")
plt.ylabel("Percentage")
plt.xlabel("Contract Type")
plt.legend(title="Churn")
plt.show()
```

people who have month to month are likely to churn out then people who are 1 and 2 years of contract

```python


# Columns you want to plot
cols = ['PhoneService', 'MultipleLines', 'InternetService',
        'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
        'TechSupport', 'StreamingTV', 'StreamingMovies']

# Set figure size (3 rows Ã— 3 cols since you have 9 features)
fig, axes = plt.subplots(3, 3, figsize=(15, 12))
axes = axes.flatten()  # flatten for easy iteration

# Loop through each column and create countplot
for i, col in enumerate(cols):
    sns.countplot(x=col, data=df, hue="Churn", ax=axes[i])
    axes[i].set_title(f"Count of Customers by {col}")
    axes[i].set_xlabel(col)
    axes[i].set_ylabel("Count")
    # Add count labels on bars
    for container in axes[i].containers:
        axes[i].bar_label(container, fontsize=8)

# Adjust layout so plots donâ€™t overlap
plt.tight_layout()
plt.show()
```

Phone Service: Customers without phone service churn more than those with it.

Multiple Lines: Customers with multiple lines churn slightly more than those with a single line.

Internet Service: Fiber optic users churn the most, while DSL users churn less.

Online Security: Churn is higher among customers without online security.

Online Backup: Customers without online backup are more likely to churn.

Device Protection: Churn is higher when customers donâ€™t have device protection.

Tech Support: Lack of tech support is linked to higher churn rates.

Streaming TV: Customers using streaming TV services show higher churn.

Streaming Movies: Churn is greater among customers with streaming movie services.
